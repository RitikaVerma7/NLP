### Top Headlines

```python
import pandas as pd
import requests
from datetime import datetime
from datetime import datetime, timedelta

# API endpoint and parameters
url = 'https://newsapi.org/v2/top-headlines?'

today = datetime.today().strftime('%Y-%m-%d')
yesterday = (datetime.today() - timedelta(days=1)).strftime('%Y-%m-%d')

parameters = {
    'from': yesterday,
    'to': today,
    'language': 'en',
    'sortBy': 'publishedAt',  # Sort by most recent
    'apiKey': NewsKey,
    'pageSize':100 # Replace with your actual API key
}

# Request data from the API
response = requests.get(url, params=parameters)
data = response.json()

# Extract articles
articles = data.get('articles', [])

# Prepare the data for the DataFrame
data_list = []
for article in articles:
    title = article.get('title', '').strip() if article.get('title') else 'Title Not Available'
    description = article.get('description', '').strip() if article.get('description') else 'Description Not Available'
    url_article = article.get('url', 'URL Not Available').strip()
    data_list.append({'Title': title, 'Description': description, 'URL': url_article})

# Create a DataFrame
df_TopNews = pd.DataFrame(data_list)

# Print the DataFrame
print(df_TopNews)
```

```python
import openai
import pandas as pd
import numpy as np
import faiss

# Set your OpenAI API key
openai.api_key = OpenAPI

# Combine 'Title' and 'Description' for embedding generation
df['combined'] = df['Title'] + " " + df['Description']

# Generate embeddings using OpenAI
def get_embeddings(text):
    response = openai.Embedding.create(
        input=text,
        model="text-embedding-ada-002"
    )
    return np.array(response['data'][0]['embedding'])

# Apply embedding generation to the dataset
df['embeddings'] = df['combined'].apply(get_embeddings)

# Convert embeddings to a numpy array for FAISS indexing
embeddings = np.vstack(df['embeddings'].values)
```

```python
# Create a FAISS index with L2 distance
index = faiss.IndexFlatL2(embeddings.shape[1])  # Dimension of embeddings
index.add(embeddings)  # Add embeddings to the index
```

```python
class LocalFAISSRetrieve:
    """
    Custom retrieval logic using FAISS.
    """
    def __init__(self, index, df):
        self.index = index
        self.df = df

    def __call__(self, query, k=5):
        # Generate embedding for the query
        query_embedding = get_embeddings(query).reshape(1, -1)

        # Search FAISS index
        distances, indices = self.index.search(query_embedding, k)

        # Retrieve the top-k results from the DataFrame
        results = self.df.iloc[indices[0]].to_dict(orient='records')
        return results
```

```python
# Initialize the custom retrieval logic
faiss_retrieve = LocalFAISSRetrieve(index=index, df=df)
```

### DSPy Setup

```python
import dspy

# Set up OpenAI GPT-3.5 Turbo as the language model
turbo = dspy.OpenAI(model='gpt-3.5-turbo', max_tokens=1000)

# Configure DSPy settings with the LM
dspy.settings.configure(lm=turbo, rm=None)  # rm is set later with the custom retrieval module
```

```python
class GenerateNews(dspy.Signature):
    """Generate personalized news with title, description, and URL."""
    context = dspy.InputField(desc="Relevant news articles matching the query")
    question = dspy.InputField(desc="User's query, e.g., 'technology news'")
    answer = dspy.OutputField(desc="Formatted list of matching news articles with title, description, and URL")
```

```python
class RAG(dspy.Module):
    def __init__(self, retrieve, num_passages=5):
        super().__init__()
        self.retrieve = retrieve
        self.num_passages = num_passages  # Correctly store num_passages as an attribute
        self.generate_answer = dspy.ChainOfThought(GenerateNews)

    def forward(self, question):
        # Retrieve relevant articles
        retrieved_articles = self.retrieve(query=question, k=self.num_passages)

        # Format retrieved articles to include title, description, and URL
        context = [
            f"Title: {article['Title']}\nDescription: {article['Description']}\nURL: {article['URL']}"
            for article in retrieved_articles
        ]

        # Generate the final answer
        prediction = self.generate_answer(context=context, question=question)
        return prediction
```

```python
# Instantiate the RAG module
rag = RAG(retrieve=faiss_retrieve, num_passages=5)
```

### Formatting the output - Professional or quick read

```python
# Define a new node for formatting news in different styles
class FormatNews(dspy.Signature):
    """Format news articles based on the preferred style."""
    context = dspy.InputField(desc="Relevant news articles matching the query")
    question = dspy.InputField(desc="User's query, e.g., 'technology news'")
    style = dspy.InputField(desc="Preferred format style, e.g., 'professional', 'quick read'")
    answer = dspy.OutputField(desc="Formatted list of news articles based on the preferred style")

# Update the RAG class to include the new node for formatting styles
class RAGWithStyle(dspy.Module):
    def __init__(self, retrieve, num_passages=5):
        super().__init__()
        self.retrieve = retrieve
        self.num_passages = num_passages
        self.generate_answer = dspy.ChainOfThought(GenerateNews)  # Existing news generation
        self.format_answer = dspy.ChainOfThought(FormatNews)  # New node for formatting

    def forward(self, question, style="professional"):
        # Retrieve relevant articles
        retrieved_articles = self.retrieve(query=question, k=self.num_passages)

        # Format retrieved articles to include title, description, and URL
        context = [
            f"Title: {article['Title']}\nDescription: {article['Description']}\nURL: {article['URL']}"
            for article in retrieved_articles
        ]

        # Generate the raw answer
        raw_answer = self.generate_answer(context=context, question=question)

        # Format the answer based on the style
        formatted_answer = self.format_answer(context=context, question=question, style=style)
        return formatted_answer

```

```python
# Instantiate the new RAG module with style support
rag_with_style = RAGWithStyle(retrieve=faiss_retrieve, num_passages=5)

# Example usage
question = "What are the latest updates in AI technology?"
style = "quick read"  # Options: 'professional', 'quick read', etc.

response = rag_with_style.forward(question=question, style=style)
print(response)
```

```python
# Example usage
question = "What are the latest updates in AI technology?"
style = "professional"  # Options: 'professional', 'quick read', etc.

response = rag_with_style.forward(question=question, style=style)
print(response)
```
